```python
plot_11 = e_data_1.copy().sort_values('Population', ascending = False)

def vulnerability (x):
    if 0>= x < 0.25:
        return "Very low"
    if  0.25>= x < 0.5:
        return "Low"
    if  x ==0.5:
        return "Mid"
    if  0.5>= x < 0.75:
        return "High"
    else:
        return "Very High"
    return
```

```python
plot_11['Overall CCVI Index']=plot_11['Overall CCVI Index'].apply(vulnerability)

fig, ax=plt.subplots(figsize=(35,10))

sns.lineplot(data=plot_11, x='states', y='Population', hue='Overall CCVI Index',
             hue_order=['Very High','High','Mid','low','Very low'],
                  ax=ax).set_title('Vulnerabilty index across states with respect to Population')

plt.show()
```
